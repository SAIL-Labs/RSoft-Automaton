```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Throughput_results.csv')
average_result = {}
figure, axes = plt.subplots(1,2,figsize=(14,6),sharex=False,sharey=False)

ax = axes[0]
for col in df.columns:
    if col == 'x':
        continue
    ax.plot(df.index,df[col], label=col)
    average_result = np.sum(df[col])/len(df[col])

ax.set_xlabel(r'Fibre length ($\mu m$)')
ax.set_ylabel('Fibre Throughput')
ax.legend()

ax = axes[1]
for col in df.columns:
    if col == 'x':
        continue
    ax.plot(col,average_result, label=col)

```
